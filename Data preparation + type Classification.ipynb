```python
from google.colab import drive
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn import svm
import pandas as pd
```

```python
# Load cleaned data
drive.mount('/content/drive')c
path = "/content/drive/MyDrive/Colab Notebooks/PokeÃÅmon Data Mining/clean_pokemon_data.csv"
pokemon_data= pd.read_csv(path, index_col=0)
```

```python
# Define feature variables
features = ['generation', 'status', 'height_m', 'weight_kg', 'abilities_number', 'ability_1', 
            'ability_2', 'ability_hidden', 'total_points', 'hp', 'attack', 'defense', 'sp_attack', 
            'sp_defense', 'speed', 'catch_rate', 'base_friendship', 'base_experience', 'growth_rate', 
            'egg_type_number', 'egg_type_1', 'egg_type_2', 'percentage_male', 'egg_cycles']

# Define target variables
targets = ['type_1', 'type_2']

# Separate features and targets
X = pokemon_data[features]
y = pokemon_data[targets]

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, columns=['generation', 'status', 'ability_1', 'ability_2', 'ability_hidden', 
                                       'growth_rate', 'egg_type_1', 'egg_type_2'])

print(X_encoded.head())
```

```python
# Print number of missing values in each column
print(pokemon_data.isnull().sum())
```

```python
# Prepare features data
X = pokemon_data[features]

# Encode categorical data
le = LabelEncoder()
for column in X.columns:
    if X[column].dtype == type(object):
        X[column] = le.fit_transform(X[column].astype(str))

# Prepare target data for type_1
y_type_1 = pokemon_data['type_1']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_type_1, test_size=0.2, random_state=42)

# Initialize classifier
clf = RandomForestClassifier()

# Fit classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate model
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# Prepare target data for type_2
y_type_2 = pokemon_data['type_2']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_type_2, test_size=0.2, random_state=42)

# Initialize classifier
clf = RandomForestClassifier()

# Fit classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate model
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_type_1, test_size=0.2, random_state=42)

# Initialize classifier
clf = svm.SVC()

# Fit classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate model
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_type_2, test_size=0.2, random_state=42)

# Initialize classifier
clf = svm.SVC()

# Fit classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate model
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
# Initialize classifiers
clf_svm = svm.SVC()
clf_rf = RandomForestClassifier()

# Split target data into two arrays
y_type_1 = y['type_1']
y_type_2 = y['type_2']

# Perform k-fold cross-validation for SVM for 'type_1'
scores_svm_type_1 = cross_val_score(clf_svm, X_encoded, y_type_1, cv=5)
print("Cross-validation scores for SVM (type_1): ", scores_svm_type_1)
print("Average cross-validation score for SVM (type_1): ", scores_svm_type_1.mean())

# Perform k-fold cross-validation for SVM for 'type_2'
scores_svm_type_2 = cross_val_score(clf_svm, X_encoded, y_type_2, cv=5)
print("Cross-validation scores for SVM (type_2): ", scores_svm_type_2)
print("Average cross-validation score for SVM (type_2): ", scores_svm_type_2.mean())

# Repeat process for Random Forest
scores_rf_type_1 = cross_val_score(clf_rf, X_encoded, y_type_1, cv=5)
print("Cross-validation scores for Random Forest (type_1): ", scores_rf_type_1)
print("Average cross-validation score for Random Forest (type_1): ", scores_rf_type_1.mean())

scores_rf_type_2 = cross_val_score(clf_rf, X_encoded, y_type_2, cv=5)
print("Cross-validation scores for Random Forest (type_2): ", scores_rf_type_2)
print("Average cross-validation score for Random Forest (type_2): ", scores_rf_type_2.mean())
```
