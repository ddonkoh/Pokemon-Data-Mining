**Impute missing values with 'none' and mean**

```python
# Replace missing values in the  columns with 'None'
pokemon_data["type_2"].fillna("None", inplace=True)
pokemon_data["ability_1"].fillna("None", inplace=True)
pokemon_data["ability_2"].fillna("None", inplace=True)
pokemon_data["ability_hidden"].fillna("None", inplace=True)
pokemon_data["egg_type_1"].fillna("None", inplace=True)
pokemon_data["egg_type_2"].fillna("None", inplace=True)
pokemon_data["growth_rate"].fillna("None", inplace=True)
```

```python
# Impute missing values in numeric columns with the mean value
pokemon_data['catch_rate'].fillna(pokemon_data['catch_rate'].mean(), inplace=True)
pokemon_data['base_friendship'].fillna(pokemon_data['base_friendship'].mean(), inplace=True)
pokemon_data['base_experience'].fillna(pokemon_data['base_experience'].mean(), inplace=True)
pokemon_data['percentage_male'].fillna(pokemon_data['percentage_male'].mean(), inplace=True)
pokemon_data['egg_cycles'].fillna(pokemon_data['egg_cycles'].mean(), inplace=True)
pokemon_data['weight_kg'].fillna(pokemon_data['weight_kg'].mean(), inplace=True)
```

```python
# Check for missing values in each column
missing_values = pokemon_data.isnull().mean() * 100

# Print the number of missing values in each column
print(missing_values)
```

**View Outliers in Numerical Variables**

```python
import numpy as np
import matplotlib.pyplot as plt

# Select only columns with numeric data type
numeric_cols = pokemon_data.select_dtypes(include=np.number).columns.tolist()

for col in numeric_cols:
  
  col_data = pokemon_data[col]

  # Calculate z-scores
  z_scores = np.abs((col_data - np.mean(col_data)) / np.std(col_data))

  # Define a threshold for outlier detection
  threshold = 3

  # Identify outliers
  outliers = np.where(z_scores > threshold)

  # Show the outliers
  if len(outliers[0]) > 0:
    print("Outliers in", col, ":", col_data.iloc[outliers])

  # Visualize distribution with box plot
  plt.boxplot(col_data)
  plt.title(col)
  plt.show()

  # Visualize z-scores with scatter plot
  plt.scatter(range(len(col_data)), col_data, c=z_scores, cmap='coolwarm')
  plt.title(col)
  plt.colorbar()
  plt.show()
```
